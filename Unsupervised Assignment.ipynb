```python
X = wine_2
K = 3
Centroids = (X.sample(n=K))
plt.scatter(X['Alcohol'],X['Color_Intensity'],c='green')
plt.scatter(Centroids['Alcohol'],Centroids['Color_Intensity'],c='blue')
plt.xlabel('Alcohol')
plt.ylabel('Color_Intensity')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
K = 3
kmeans = KMeans(n_clusters=K, random_state=42)
X['Cluster'] = kmeans.fit_predict(X[['Alcohol','Color_Intensity']])
```

```python
X.head(10)
```

```python
centroids = kmeans.cluster_centers_
sns.scatterplot(X, x='Alcohol', y='Color_Intensity', hue='Cluster', palette='Set1')
plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='x', s=50)
plt.xlabel('Alcohol')
plt.ylabel('Color_Intensity')
plt.show()
```

```python
inertia = kmeans.inertia_
print('Inertia:',inertia)
```

```python
K_range = range(1, 11)
inertia_values = []
for K in K_range:
    kmeans = KMeans(n_clusters=K, random_state=42)
    kmeans.fit(X[['Alcohol', 'Color_Intensity']])
    inertia = kmeans.inertia_
    inertia_values.append(inertia)   
plt.plot(K_range, inertia_values, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_avg = silhouette_score(X[['Alcohol', 'Color_Intensity']], X['Cluster'])
print('Silhouette Score:', silhouette_avg)
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Alcohol')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(13,18))
ax=dendrogram(X,orientation="right",labels=np.array(X['Alcohol']),leaf_rotation=25,leaf_font_size=15)
plt.tight_layout()
plt.show()
```
